```python
import numpy as np
import matplotlib.pyplot as plt

# Generate noisy 2D data
np.random.seed(0)
data = np.vstack([
    np.random.randn(50,2) + [2,2],
    np.random.randn(50,2) + [-2,-2],
    np.random.randn(50,2) + [2,-2]
])

# Initialize neurons
num_neurons = 3
weights = np.random.randn(num_neurons, 2)

lr = 0.1
epochs = 50

for epoch in range(epochs):
    for x in data:
        # Find winner neuron
        distances = np.linalg.norm(weights - x, axis=1)
        winner = np.argmin(distances)

        # Update only winner
        weights[winner] += lr * (x - weights[winner])

# Plot results
for i in range(num_neurons):
    plt.scatter(data[:,0], data[:,1])
    plt.scatter(weights[i][0], weights[i][1], marker='X', s=200)

plt.title("Competitive Learning Output")
plt.show()
```
